```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from scipy.stats import kruskal,chisquare,chi2_contingency
from imblearn.over_sampling import SMOTE
from catboost import CatBoostClassifier
import funcs
from sklearn.metrics import roc_curve, auc, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.mixture import GaussianMixture
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import KMeans
import itertools
import warnings
warnings.filterwarnings('ignore') # for Kmeans memory leak warning
```

```python
data = pd.read_csv("../../adni/objects/adni_df.csv", index_col=0)
data.head()
```

```python
df = data.copy()
```

```python
df = funcs.preproc_df(df)
```

## Filter-based feature selection

### KW test for continuous vars & Kramer's V for categorical vars

```python
cat_features = [col for col in df.columns if (
    col.startswith("NX") or
    col.startswith("PX") or
    col.startswith("PT") or
    col.startswith("MH") or
    (col.startswith("GD") and col != "GDTOTAL") or
    col in ["mother", "father", "HMSCORE", "apoe"]
)]
```

```python
cont_features = [col for col in df.columns if 
                 col not in cat_features and 
                col != "DIAGNOSIS"]
```

```python
kr_stat = []
kr_p = []
for f in cont_features:
    groups = [df[f][df['DIAGNOSIS'] == cat] for cat in df['DIAGNOSIS'].unique()]
    kruskal_result = kruskal(*groups)
    kr_stat.append(kruskal_result.statistic)
    kr_p.append(kruskal_result.pvalue)
```

```python
result_kw = pd.DataFrame({"feature": cont_features,
                         "stat": kr_stat,
                         "pvalue": kr_p})
```

```python
alpha=0.05
for f in cont_features:
    groups = [df[f][df['DIAGNOSIS'] == cat] for cat in df['DIAGNOSIS'].unique()]
    
    # Perform omnibus Kruskal-Wallis test
    kruskal_result = kruskal(*groups)
    
    # Store the omnibus test result
    kr_stat.append(kruskal_result.statistic)
    kr_p.append(kruskal_result.pvalue)
    
    # Perform pairwise Kruskal-Wallis tests between all combinations of categories in DIAGNOSIS
    pairwise_significant = True
    for cat1, cat2 in itertools.combinations(df['DIAGNOSIS'].unique(), 2):
        group1 = df[f][df['DIAGNOSIS'] == cat1]
        group2 = df[f][df['DIAGNOSIS'] == cat2]
        
        # Perform pairwise Kruskal-Wallis test
        pairwise_result = kruskal(group1, group2)
        
        # Check if p-value is below the threshold
        if pairwise_result.pvalue >= alpha:
            pairwise_significant = False
            break  # If any pairwise comparison is not significant, stop checking

    # Only append to results if all pairwise comparisons are significant
    if pairwise_significant:
        print(f"Feature {f} passed all pairwise comparisons with p < {alpha}")
```

### Cont features significant across AD, MCI and HC at p = 0.01

```python
result_kw[result_kw['pvalue'] < 0.01]
```

```python
cont_features_sig = result_kw[result_kw['pvalue'] < 0.01].feature.tolist()
```

```python
cr = []

for f in cat_features:
    cont_tab = pd.crosstab(index=data['DIAGNOSIS'], columns=data[f])
    chi2, p, dof, expected = chi2_contingency(cont_tab)
    n = cont_tab.sum().sum()
    phi2 = chi2 / n
    r, k = cont_tab.shape
    cramers_v = np.sqrt(phi2 / min(r-1, k-1))
    cr.append(cramers_v)
```

```python
result_kramer = pd.DataFrame({"feature": cat_features,
                         "Kramer": cr})
result_kramer = result_kramer.sort_values(by="Kramer", ascending=False)
```

### Cat features with at least moderate association with the outcome

```python
result_kramer[result_kramer['Kramer'] >= 0.11]
```

```python
cat_features_sig = result_kramer[result_kramer['Kramer'] >= 0.11].feature.tolist()
```

```python
feat = cont_features_sig + cat_features_sig
```

### Model after feature selection

```python
outcome = df['DIAGNOSIS']
outcome = outcome.replace({"Control": 0,
                  "MCI": 1,
                  "AZ": 2})
```

```python
df.drop('DIAGNOSIS', axis = 1, inplace=True)
```

```python
X, X_test, y, y_test = train_test_split(df[feat], outcome, test_size=0.2, random_state=43)
```

```python
smote_over = SMOTE(random_state=44)
X, y = smote_over.fit_resample(X, y)
```

```python
best_params_feat_selection = {'iterations': 1800, 
                              'learning_rate': 0.2736095674018922, 
                              'l2_leaf_reg': 3.531079234474505, 
                              'bagging_temperature': 0.571547709531307, 
                              'random_strength': 1.5028953428173903, 
                              'depth': 9, 
                              'min_data_in_leaf': 79,
                              'random_seed': 44,
                              'loss_function': "MultiClass",
                              'colsample_bylevel': 0.9003486064573871}
```

```python
final_model = CatBoostClassifier(**best_params_feat_selection, verbose=False, cat_features = cat_features_sig)
```

```python
final_model.fit(X, y)
```

```python
predictions = final_model.predict(X_test)
predictions_proba = final_model.predict_proba(X_test)
```

```python
funcs.metrics_merged(y_test, predictions, predictions_proba)
```

```python
confusion_matrix(y_test, predictions)
```

```python
n_classes = len(np.unique(y_test))
class_labels = ["Control", "MCI", "AD"]
colors = ['blue', 'red', 'green']  

plt.figure(figsize=(10, 8))

for i in range(n_classes):
    fpr, tpr, _ = roc_curve(y_test == i, predictions_proba[:, i]) 
    roc_auc = auc(fpr, tpr)  
    plt.plot(fpr, tpr, color=colors[i], label=f'{class_labels[i]} (AUC = {roc_auc:.2f})')


plt.plot([0, 1], [0, 1], 'k--', label='Random Guessing')  
plt.xlim([-0.05, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('AUROC Curves for Multi-class model')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

## PCA

```python
data_feat = df[feat]
```

```python
pipeline = Pipeline([("scaler", StandardScaler()), ("pca", PCA(n_components=2)),])
```

```python
pca_data = pd.DataFrame(
    pipeline.fit_transform(data_feat),
    columns=["PC1", "PC2"],
    index=data_feat.index
)
pca_data['outcome'] = outcome.replace({0: "Control", 1:"MCI", 2:"AZ"})
pca_step = pipeline.steps[1][1]
```

### MCI overlaps with both Control and AD

```python
g = sns.scatterplot(data=pca_data, x="PC1", y="PC2", hue="outcome")


g.set_xlabel(f"PC1 ({pca_step.explained_variance_ratio_[0]*100:.2f} %)")
g.set_ylabel(f"PC2 ({pca_step.explained_variance_ratio_[1]*100:.2f} %)")

#plt.savefig("../pics/PCA_with_loadings.tiff", dpi=300)
plt.show()
```

```python
pca_MCI = pca_data[pca_data["outcome"] == "MCI"]
```

### Based on selected features the optimal number of clusters in MCI group is three

```python
inertias = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(pca_MCI.iloc[:,0:2])
    inertias.append(kmeans.inertia_)

# Plot inertia to find the elbow
plt.plot(range(1, 10), inertias, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
#plt.savefig("../pics/Inertia.tiff", dpi=300)
plt.show()
```

```python
gmm = GaussianMixture(n_components=3, covariance_type='tied')
labels = gmm.fit_predict(pca_MCI.iloc[:,0:2])
```

```python
labels_MCI = np.where(labels == 0, "MCI1", np.where(labels == 1, "MCI2", "MCI3"))
```

```python
pca_data["cluster"] = pca_data["outcome"]
pca_data.loc[pca_data["outcome"] == "MCI", "cluster"] = labels_MCI
```

```python
g = sns.scatterplot(data=pca_data, x="PC1", y="PC2", hue="cluster")


# Add variance explained by the
g.set_xlabel(f"PC1 ({pca_step.explained_variance_ratio_[0]*100:.2f} %)")
g.set_ylabel(f"PC2 ({pca_step.explained_variance_ratio_[1]*100:.2f} %)")

#plt.savefig("../pics/PCA_with_loadings_2.tiff", dpi=300)
plt.show()
```

### Clusters can be characterized by proximity to either HC or AD groups

```python
grouped_data = pca_data.groupby('cluster')[['PC1', 'PC2']].mean()
Z = linkage(grouped_data, method='ward')

# Create a dendrogram to visualize the clustering
plt.figure(figsize=(10, 7))
dendrogram(Z, labels=grouped_data.index)  # 'outcome' includes AZ, Control, MCI1, MCI2, MCI3
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
#plt.savefig("../pics/dendrogram.tiff", dpi=300)
plt.show()
```
